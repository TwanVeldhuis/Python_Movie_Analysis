```python
import requests
import pandas as pd
import numpy as np

api_key = '<HERE IS WHERE YOU HAVE TO PUT IN YOUR THEMOVIEDB.ORG API KEY>'
```

```python
for id in range(550, 556):
    url = 'https://api.themoviedb.org/3/movie/{}?api_key={}'.format(id, api_key[0])
    response = requests.get(url)

    row = pd.DataFrame.from_dict(response.json(), orient='index').transpose()
    
    if id == 550:
        df = row
    else:
        df = pd.concat([df, row], axis=0)

df = df.reset_index(drop=True)
df
```

```python
movies = df[["budget", "genres", "id", "imdb_id", "original_title", "release_date", "revenue", "runtime"]]
movies
```

```python
genres = movies.explode('genres').reset_index(drop=True)
only_genres = genres['genres'].apply(pd.Series).drop_duplicates()

only_genres.to_csv('genres.csv', index=False)
only_genres
```

```python
genres = movies.explode('genres').reset_index(drop=True)
genres_and_ids = pd.concat([genres['id'], genres['genres'].apply(pd.Series).drop('id', axis=1)], axis=1)
```

```python
genres_list = only_genres['name'].values.tolist()
movies_expanded = pd.DataFrame(movies.drop('genres', axis=1))

for genre in genres_list:
    movies_expanded[genre] = [0,0,0,0,0,0]

movies_expanded
```

```python
for index, row in genres_and_ids.iterrows():
    index = 0
    for results_index, results_row in movies_expanded.iterrows():
        if row[0] == results_row[1]:
            movies_expanded[row[1]][index] = 1
        index = index + 1
            
movies_expanded.to_csv('movies.csv', index=False)
movies_expanded
```

```python
movies_expanded['release_date'] =  pd.to_datetime(movies_expanded['release_date'])
movies_expanded

datetimes = pd.DataFrame(movies_expanded[['id', 'release_date']])
datetimes['day'] = datetimes.release_date.dt.day
datetimes['month'] = datetimes.release_date.dt.month
datetimes['year'] = datetimes.release_date.dt.year
datetimes['day_of_week'] = datetimes.release_date.dt.day_name()
datetimes = datetimes.reset_index(drop=True)

datetimes.to_csv('datetimes.csv', index=False)
datetimes
```
